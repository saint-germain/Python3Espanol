En este notebook jugaremos con los niveles de una imagen reducida de Fomalhaut para encontrar al exoplaneta Fomalhaut b.
```python
from astropy.io import fits
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n=fits.open('PCA_FOMAL_full_centroids.fits')
```

```python
dst=n[0].data
```

```python
dst
```

```python
plt.imshow(dst)
```

```python
dst.min()
```

```python
dst.max()
```

```python
ninf=0.
nsup=0.005
```

```python
plt.figure(figsize=(10,10))
plt.imshow(dst*(dst>ninf)*(dst<nsup),cmap="gist_heat")
plt.scatter(1173,1205,s=300,facecolors='none', edgecolors='b',lw=2)
plt.xlim(900,1250)
plt.ylim(1100,1400)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(dst*(dst>-0.0045)*(dst<0.004),cmap="gist_heat")
plt.scatter(1173,1205,s=300,facecolors='none', edgecolors='b',lw=2)
plt.xlim(900,1250)
plt.ylim(1100,1400)
```
